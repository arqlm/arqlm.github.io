1.   Analizar (mediante herramientas estadísticas simples) la cantidad, la calidad y la ubicación de los datos disponibles.
2.   Definir la(s) zona(s) de estudio. Una división del campo en varias sub-zonas puede ser relevante si uno observa cambios abruptos en la distribución espacial de valores, o si la geología del fenómeno lo indica.
3.   Anticipar dificultades o problemas que puedan surgir en la fase de estimación local (por ejemplo, presencia de valores atípicos que se destacan de aquellos de los datos vecinos).
